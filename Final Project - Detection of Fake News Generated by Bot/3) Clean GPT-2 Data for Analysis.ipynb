```python
real_df = pd.DataFrame(data[['titles', 'authors', 'url', 'source', 'dates', 'category', 'gpt2_inputs', 'real_news']])
real_df['is_fake'] = 0
real_df['gpt2_inputs'] = np.nan
real_df.rename(columns={'real_news': 'news'}, inplace= True)
print(real_df.shape)
real_df
```

```python
output = pd.concat([real_df, fake_df], axis = 0)
output.reset_index(drop= True, inplace= True)
print(output.shape)
output.to_csv('data_cleaned_50words.csv', index= False)
```
