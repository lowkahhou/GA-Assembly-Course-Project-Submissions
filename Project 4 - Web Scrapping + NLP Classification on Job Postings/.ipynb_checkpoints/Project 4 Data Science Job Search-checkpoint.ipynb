# Data Preprocessing

```python
df = pd.read_csv('Job_Data_main.csv', index_col= 0)
```

```python
# to remove unnecessary 1st two and last two characters (i.e. []') in all columns except urls  

for col in [index for index in df.columns if index != 'urls']: # exclude column "urls"
    df[col] = [i[2:-2] for i in df[col]]
```

```python
# replace empty cell with np.nan

df.replace("", np.nan, inplace= True)
```

```python
# Check for empty cells

df.isna().sum()
```

```python
# Replace with 'Undisclosed'

df['address'] = df['address'].replace(np.nan, 'Undisclosed')
df['seniority'] = df['seniority'].replace(np.nan, 'Undisclosed')
df['req'] = df['req'].replace(np.nan, 'Undisclosed')
```

```python
# Drop rows with empty salary
# Do not intend to impute missing values for salary which is our target
# as wrong assumptions will greatly affect our insights

drop_index = list(df[df['salary'].isna()].index)
df = df.drop(df.index[drop_index])
```

```python
# Drop duplicate entries, may have different url links but same job

cols_check_dup = ['company', 'job_title', 'address', 'emp_type', 'seniority', 'job_cat', 'salary', 'req']
df.drop_duplicates(cols_check_dup, inplace= True)
```

```python
# Split salary range to min and max

df['min_salary'] = [i.split()[0] for i in df['salary']]
df['max_salary'] = [i.split()[1] for i in df['salary']]
df['min_salary'] = [int(i.replace('$', '').replace(',', '').replace("'", "")) for i in df['min_salary']]
df['max_salary'] = [int(i.replace('$', '').replace(',', '').replace("'", "")) for i in df['max_salary']]
```

```python
# Check for typo in minimum salary, should be $5500 instead of $550

df[df['min_salary'] < 1000]
```

```python
# To correct the typo in minimum salary

df['min_salary'] = df['min_salary'].replace(550, 5500)
```

```python
# Some salaries are given as annual, to convert to salary PER MONTH

index_annual = df[df['salary_type'] == 'Annually'].index

df.loc[index_annual, 'min_salary'] = round(df.loc[index_annual, 'min_salary']/12)
df.loc[index_annual, 'max_salary'] = round(df.loc[index_annual, 'max_salary']/12)
```

```python
# To identify typos by finding jobs with very big salary range 
# These should be typos in the maximum salary, have additional zero

df[df['max_salary'] - df['min_salary'] > 15000]
```

```python
# To correct the typo in maximum salary

index_max_salary = list(df[df['max_salary'] - df['min_salary'] > 15000].index)
df.loc[index_max_salary, 'max_salary'] = df.loc[index_max_salary, 'max_salary'] / 10
```

```python
# Get the average salary based on the range given

df['avg_salary'] = round(((df['max_salary'] - df['min_salary']) / 2) + df['min_salary'])
```

```python
# To categorise all the job titles as either
#['Data Scientist', 'Data Engineer', 'Data Analyst', 'Business Analyst', 'Software Engineer', 'Researcher', 
#         'Manager', 'Engineer', 'Others']

df['job_title'] = df['job_title'].str.lower()

df.loc[df[df['job_title'].str.contains('data scientist')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains('data science')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains('deep')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains('nlp')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains('language')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains('machine')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains('image')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains('video')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains('robotic')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains('ai ')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains(' ml ')].index, 'job_title'] = 'Data Scientist'
df.loc[df[df['job_title'].str.contains('algorithm')].index, 'job_title'] = 'Data Scientist'


df.loc[df[df['job_title'].str.contains('data engineer')].index, 'job_title'] = 'Data Engineer'
df.loc[df[df['job_title'].str.contains('big data')].index, 'job_title'] = 'Data Engineer'
df.loc[df[df['job_title'].str.contains('hadoop')].index, 'job_title'] = 'Data Engineer'
df.loc[df[df['job_title'].str.contains('data centre')].index, 'job_title'] = 'Data Engineer'
df.loc[df[df['job_title'].str.contains('data center')].index, 'job_title'] = 'Data Engineer'
df.loc[df[df['job_title'].str.contains('architect')].index, 'job_title'] = 'Data Engineer'
df.loc[df[df['job_title'].str.contains('cloud')].index, 'job_title'] = 'Data Engineer'
df.loc[df[df['job_title'].str.contains('warehouse')].index, 'job_title'] = 'Data Engineer'


df.loc[df[df['job_title'].str.contains('analyst')].index, 'job_title'] = 'Data Analyst'
df.loc[df[df['job_title'].str.contains('analytic')].index, 'job_title'] = 'Data Analyst'
df.loc[df[df['job_title'].str.contains('analysis')].index, 'job_title'] = 'Data Analyst'

df.loc[df[df['job_title'].str.contains('business')].index, 'job_title'] = 'Business Analyst'

df.loc[df[df['job_title'].str.contains('software')].index, 'job_title'] = 'Software Engineer'

df.loc[df[df['job_title'].str.contains('research')].index, 'job_title'] = 'Researcher'

df.loc[df[df['job_title'].str.contains('manager')].index, 'job_title'] = 'Manager'

df.loc[df[df['job_title'].str.contains('engineer')].index, 'job_title'] = 'Engineer'
```

```python
# Believe that the unstructured text data in job category, desc, req, skills and seniority have a part to play in 
# deciding the salary, hence to vectorise these text and include them for modelling.

df_words = pd.DataFrame()

for col in ['job_cat', 'desc', 'req', 'skills', 'seniority']:    
    cvec = CountVectorizer(stop_words= 'english', ngram_range=(1,2))
    temp_df = pd.DataFrame(cvec.fit_transform(df[col]).todense())
    word_name = [col + '_' + word for word in cvec.get_feature_names()]
    temp_df.columns = word_name
    df_words = pd.concat([df_words, temp_df], axis = 1)
df_words.sum().sort_values(ascending = False)[:75]

# The following words make sense in deciding the salary, to use this for modelling
```

# Chi Square Test

```python
# To find top 50 features with lowest p-values 
# (statistical difference of the feature counts that appear in high salary jobs vs low salary jobs)

chi2_df = df['avg_salary_range']
chi2_df = pd.concat([chi2_df, df_words], axis = 1)
chi2_df = chi2_df.groupby('avg_salary_range').sum()

chi2_score = []
for i in chi2_df.columns:
    _, p = chisquare([chi2_df[i][0], chi2_df[i][1]])
    chi2_score.append(p)
    
chi2_df = chi2_df.T.rename(columns= str).reset_index()
chi2_df.columns = ['features', 'low salary count', 'high salary count']
chi2_df['p_value'] = chi2_score
chi2_df.sort_values('p_value', ascending= True).head(50)

# Alot of the words do no make sense in distinguishing high salary vs low salary, 
# such as climate, regulatory compliance, products, tookitaki, etc.

# It could be due to that there are some specify high salary jobs with unique words which are not representative of all
# the high salary jobs.Hence next step is to only look at features with high counts for either low salary or high salary.
```

```python
# To find top 50 features with p_value < 0.05 and count of each feature at least more than 200 times in either low salary
#   or high salary jobs

chi2_df[((chi2_df['low salary count'] > 200) | (chi2_df['high salary count'] > 200)) & (chi2_df['p_value'] < 0.05)
       ].sort_values('p_value')
```

**Observations from Chi Square Test**

1) High salary job requires experienced (in years) personnel with strong management ability

2) Low salary job requires more of research and microsoft skills.

The results from chi square test are based on the occurence of words in a low salary vs high salay jobs.
Next, to build a classification model and find the top features that determine whethere low salary or high salary.

# Features Selection

```python
# Using Select KBest (f-score) to see if the top few features make sense 
# Infeasible to run RFECV with so many columns as it take very long
```

```python
df_X = df[['job_title','emp_type']]
df_X = pd.get_dummies(df_X, drop_first= True) # To get dummies for job_title and emp_type

# And then to merge with the vectorised words from job_category, description, requirement, skills and seniority
df_X = df_X.merge(df_words, left_on= df_X.index, right_on= df_words.index, how = 'left') 
df_X.drop(columns= 'key_0', inplace= True)

feature_names = df_X.columns # Save feature names for later use

sscaler = StandardScaler()
X = sscaler.fit_transform(df_X)

y = df['avg_salary_range']
```

```python
print(X.shape, '\n', y.shape)
```

```python
# Using selectkbest to pick top best features

skb_f = SelectKBest(f_classif)

# Train the selector on the data.
skb_f.fit(X, y)
skb_f_scores = pd.DataFrame({'features': feature_names, 'f_scores': skb_f.scores_})
skb_f_scores.sort_values(by= 'f_scores', inplace= True, ascending= False)
skb_cols = list(skb_f_scores['features'])
```

```python
skb_cols[:5000]
```

# Model Selection

```python
y = df['avg_salary_range']
y = y.astype(int)
y.value_counts()
```

```python
print('baseline accuracy =', 1- y.mean())
```

```python
# Model 1a: Logisitic Regression (Ridge)
# Try out different number of top features from SelectKBest

for i in np.arange(1000, 8000, 1000):

    X = df_X[skb_cols[:i]]
    X = X.astype(float)
    feature_names = X.columns
    sscaler = StandardScaler()
    X = sscaler.fit_transform(X)

    y = df['avg_salary_range']

    logreg = LogisticRegression(C = 1.0, random_state= 42, penalty= 'l2', solver= 'liblinear')

    score = cross_val_score(logreg, X, y, cv = 5)
    print(i, 'features scores', '\n', score, '\n', np.mean(score), '\n')
    
# 7000 features got quite good accuracy score
```

```python
# Model 1b: Logistic Regression (Rdige) with 7000 features and optimised C = 0.3

X = df_X[skb_cols[:7000]]
X = X.astype(float)
sscaler = StandardScaler()
X = sscaler.fit_transform(X)

y = df['avg_salary_range']

logreg = LogisticRegression(C = 0.3, random_state= 42, penalty= 'l2', solver= 'sag', n_jobs= -1)

score = cross_val_score(logreg, X, y, cv = 5)
print(score, '\n', np.mean(score))

# Good accuracy score
```

```python
# Model 2: Logistic Regression (Lasso)

for i in np.arange(1000, 8000, 1000):

    X = df_X[skb_cols[:i]]
    X = X.astype(float)
    feature_names = X.columns
    sscaler = StandardScaler()
    X = sscaler.fit_transform(X)

    y = df['avg_salary_range']

    logreg = LogisticRegression(C = 1.0, random_state= 42, penalty= 'l1', solver= 'liblinear')

    score = cross_val_score(logreg, X, y, cv = 5)
    print(i, 'features scores', '\n', score, '\n', np.mean(score), '\n')
```

```python
# Model 3: Naive Bayes Multinomial

for i in np.arange(1000, 8000, 1000):

    X = df_X[skb_cols[:i]]
    X = X.astype(float)
    
    y = df['avg_salary_range']

    nbm = MultinomialNB()

    score = cross_val_score(nbm, X, y, cv = 5)
    print(i, 'features scores', '\n', score, '\n', np.mean(score), '\n')
```

```python
# Model 3: SVM Classifier

X = df_X[skb_cols[:7000]]
X = X.astype(float)
X = X.values
y = df['avg_salary_range']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify = y)

# Only linear kernel can return feature coef, however since objective now is not to find important feature, hence will
# try other kernel

svc = SVC(C = 1, gamma= 0.0001, kernel= 'rbf', random_state= 42, probability= True)
svc.fit(X_train, y_train)
y_hat = svc.predict(X_test)
pp_svc = svc.predict_proba(X_test)

svc_auc = roc_auc_score(y_test, y_hat)
svc_confusion = confusion_matrix(y_test, y_hat)

print(svc_auc)
print(svc_confusion)

# Low False Positive Rate
# only predict 3 high salary wrongly
```

```python
# Model 4: Random Forest Classifier

X = df_X[skb_cols[:7000]]
X = X.astype(float)
X = X.values
y = df['avg_salary_range']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, stratify = y)

rfc = RandomForestClassifier(n_estimators= 500, random_state= 42, n_jobs= -1) 
rfc.fit(X_train, y_train)
y_hat = rfc.predict(X_test)
pp_rfc = rfc.predict_proba(X_test)

rfc_auc = roc_auc_score(y_test, y_hat)
rfc_confusion = confusion_matrix(y_test, y_hat)

print(rfc_auc)
print(rfc_confusion)

# Predicts 25 high salary wrongly 
```
